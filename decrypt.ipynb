```python
#Libraries and installs that used
!pip install pycrypto
from Crypto.Cipher import AES 

cipher_file = '/content/ciphertext.dat'  # File contain ciphertext
key_file="/content/key.dat"              # File contain key

key= open(key_file, 'rb').read() #Open and read key file

ciphered_data = open(cipher_file, 'rb').read() #Open and read ciphertext file

cipher = AES.new(key, AES.MODE_ECB)            #Use ECB mode to decrypt ciphertext file
original_data = cipher.decrypt(ciphered_data) 

with open('/content/plaintext.dat', 'wb') as decrypt_file: #To store the decryption text in a file
    decrypt_file.write(original_data)
```
