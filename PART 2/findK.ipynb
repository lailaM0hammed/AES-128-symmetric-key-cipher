```python
"""Python libraries and installiziation to decrypt an encrypted message"""
!pip install pycrypto
import itertools 
from Crypto.Cipher import AES
import timeit
''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''''

start = timeit.default_timer() # start run time Timer to find the correct key

#open and read dat files in bytes
ciphertext = open("/content/ciphertext2.dat" , "rb").read() #encrypted text 
partialkey = open("/content/partial-key.dat" , "rb").read() #96-bits of the key
 
''' nested iteration to find the missing last 4 bytes of the key''' 
for a, b, c, d in itertools.product(range(255), repeat = 4):
    # Concatenate the partial key with nested iteration  random values
    key = partialkey  + (chr(a) + chr(b) +chr(c) + chr(d)).encode('latin-1')
    # Decrypt the ciphertext using ECB mode with guessed key 
    cipher =AES.new(key, AES.MODE_ECB)
    decryptedtext= cipher.decrypt(ciphertext)
    #Checking if the decrypted text is found  
    if(decryptedtext[:5] == b'Salam'):
      #Simple print 
      print("- The missing secret key :",((chr(a) + chr(b) +chr(c) + chr(d)).encode('latin-1')).hex() )
      print("- the decrypted message : ", decryptedtext.decode())
      #Exit the iteration
      break 
stop = timeit.default_timer()# stop run time Timer to find the correct key
print('- run Time: ', stop - start ) #simple print
```
